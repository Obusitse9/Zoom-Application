## Check if Accident Has Occured

```python
# Returns the point at which the signal peaks above a certain threshold
def getPeakAtThreshold(signal,threshold):
    return find_peaks(signal,height=threshold)
```

```python
def wasDriving(speed_signal,time,speed_threshold):                            # Returns true if the user was driving in the last 5 seconds of the potential accident
    # IF 5 seconds has not yet passed
    low_index = 0
    upper_index=0
    if(time<25):
        low_index=0
        upper_index=time
    else:
        low_index=time-25
        upper_index = time
    
    for i in range(low_index,upper_index):
        if speed_signal[i]>speed_threshold: return True
    
    return False
```

```python
threshold = 30                                  # Acceleration threshold for accident, measured in m/s^2
drving_speed_threshold = 35                     # Threshold for which the user is considered to be driving, measured in km/h
n = getPeakAtThreshold(z_data,threshold)        # Check at what time accident may have occured

Accident = False
if (len(n[0])>0):     
    if (wasDriving(speed,n[0][0],drving_speed_threshold)):
        Accident = True
        print("An accident has occured! ")   
else: print("There has not been a accident. ")                      
```
